## **Propuesta de Cambios en las Dimensiones y Hechos**

---

### **Dim_Fecha**
**Fuente de datos**: nac2017, nac2018, nac2021, nac2022

**Campos**:
- ANO (Año de ocurrencia del nacimiento)
- MES (Mes de la ocurrencia del nacimiento)

**Transformaciones**:
1. Renombrar `ANO` a `anio`.
2. Agregar una columna `trimestre` calculada en base a `MES`.
3. Crear la columna `idFecha` combinando `anio` y `mes` en formato `YYYY-MM`.

**Destino de datos**: `Dim_Fecha`

**Campos resultantes**:
- idFecha
- mes
- trimestre
- anio

---

### **Dim_Madre**
**Fuente de datos**: nac2017, nac2018, nac2021, nac2022

**Campos**:
- EDAD_MADRE (Edad de la madre a la fecha del parto)
- EST_CIVM (Estado conyugal de la madre)
- NIV_EDUM (Último nivel de estudio de la madre)
- ULTCURMAD (Último año o grado aprobado por la madre)
- SEG_SOCIAL (Régimen de seguridad social de la madre)

**Transformaciones**:
1. Renombrar `EDAD_MADRE` a `rango_edad`.
2. Renombrar `EST_CIVM` a `estado_conyugal`.
3. Renombrar `NIV_EDUM` a `maximo_nivel_educativo`.
4. Renombrar `ULTCURMAD` a `maximo_año_o_grado_aprobado`.
5. Renombrar `SEG_SOCIAL` a `regimen_seguridad_social`.
6. Reemplazar los valores `0` en `maximo_año_o_grado_aprobado` con `99`.
7. Transformar los valores categóricos en su correspondiente descripción (por ejemplo, 1 = “Contributivo”, etc.).

**Destino de datos**: `Dim_Madre`

**Campos resultantes**:
- rango_edad
- estado_conyugal
- maximo_nivel_educativo
- maximo_año_o_grado_aprobado
- regimen_seguridad_social

---

### **Dim_Hijo**
**Fuente de datos**: nac2017, nac2018, nac2021, nac2022

**Campos**:
- SEXO (Sexo del nacido vivo)
- PESO_NAC (Peso al nacer)

**Transformaciones**:
1. Renombrar `PESO_NAC` a `peso`.
2. Agregar la columna `talla` y establecer su valor como `null` (vacío).
3. Reemplazar los valores nulos en `peso` con `9`.
4. Convertir los valores categóricos en descripciones.

**Destino de datos**: `Dim_Hijo`

**Campos resultantes**:
- sexo
- peso
- talla

---

### **Dim_Parto**
**Fuente de datos**: nac2017, nac2018, nac2021, nac2022

**Campos**:
- TIPO_PARTO (Tipo de parto)
- MUL_PARTO (Multiplicidad del embarazo)

**Transformaciones**:
1. Renombrar `TIPO_PARTO` a `tipo_parto`.
2. Renombrar `MUL_PARTO` a `multiplicidad_embarazo`.
3. Reemplazar los valores categóricos por su descripción correspondiente.
4. Reemplazar los valores `4` en `TIPO_PARTO` con `9` (sin información).

**Destino de datos**: `Dim_Parto`

**Campos resultantes**:
- tipo_parto
- multiplicidad_embarazo

---

### **Dim_Atencion_Parto**
**Fuente de datos**: nac2017, nac2018, nac2021, nac2022

**Campos**:
- SIT_PARTO (Sitio de parto)
- ATEN_PAR (Atención del parto)
- PROFESION (Profesión de quien certifica el parto)

**Transformaciones**:
1. Renombrar `SIT_PARTO` a `sitio_parto`.
2. Renombrar `ATEN_PAR` a `atendido_por`.
3. Renombrar `PROFESION` a `profesion_certificadora`.
4. Reemplazar valores categóricos por descripciones textuales (por ejemplo, 1 = “Médico”).

**Destino de datos**: `Dim_Atencion_Parto`

**Campos resultantes**:
- sitio_parto
- atendido_por
- profesion_certificadora

### **Hecho_Nacimiento**

- N_HIJOSV (Número de hijos nacidos vivos)
