Las variables candidatas descritas en el informe son las siguientes:

COD_DPTO: Departamento donde ocurrió la defunción 

COD_MUNIC: Municipio donde ocurrió la defunción 

A_DEFUN: Área donde ocurrió la defunción (Cabecera municipal, centro poblado, rural disperso) 

SIT_DEFUN: Sitio donde ocurrió la defunción (Hospital/clínica, centro/puesto de salud, casa/domicilio, lugar de trabajo, via pública) 

ANO: Año en que ocurrió la defunción 

MES: Mes en que ocurrió la defunción 

PMAN_MUER: Probable manera de muerte (Natural, violenta, en estudio) 

MAN_MUER: Probable manera de muerte (violenta) (Suicidio. Homicidio, accidente de tránsito) 

C_BAS1: Código de la causa Básica de la defunción 

MU_PARTO: La muerte fetal o del menor de un año ocurrió con relación al parto (Antes, durante, después) 

T_PARTO: Tipo de parto - el feto o el menor de un año fallecido nació por parto (Espontáneo, cesárea, instrumentado) 

TIPO_EMB: Multiplicidad - el feto o el menor de un año fallecido nació de un embarazo (Simple, doble, triple, cuádruple o más) 

CONS_EXP: Certificado de defunción expedido por (Médico tratante, médico no tratante, médico legista, personal de salud autorizado, funcionario del registro civil) 

IDPROFCER: Profesión de quien certifica la defunción (Médico, enfermero(a), auxiliar de enfermería, promotor(a) de salud, funcionario del registro civil, médico legista) 

N_HIJOSV: Número de hijos nacidos vivos de la madre (incluyendo el presente) 

N_HIJOSM: Número de hijos nacidos muertos de la madre (incluyendo el presente) 
 
EDAD_MADRE: Edad de la madre del feto o del menor de un año fallecido 

CODPTORE: Departamento de residencia habitual de la madre 

CODMUNRE: Municipio de residencia habitual de la madre 

AREA_RES: Área de residencia habitual del fallecido (para muerte fetal o de menor de un año la el de la madre) 

SEG_SOCIAL: Régimen de seguridad social del fallecido (para muerte fetal, o de menor de un año el de la madre) 

EST_CIVM: Estado conyugal de la madre 

NIV_EDUM: Ultimo nivel de estudio que aprobó la madre 

ULTCURMAD: Último año o grado aprobado de la madre 

ASIS_MED: ¿Recibió asistencia médica durante el proceso que lo llevo a la muerte? 

SEXO: Sexo del fallecido 

T_GES: Tiempo de gestación 

T_GES_AGRU_CIE: Tiempo de gestación ajustado a la agrupación sugerida por la CIE 

PESO_NAC: Peso al nacer del feto o del menor de un año 

```python
variablesABuscar = set([
"COD_DPTO",
"COD_MUNIC",
"A_DEFUN",
"SIT_DEFUN",
"ANO",
"MES",
"PMAN_MUER",
"MAN_MUER",
"C_BAS1",
"MU_PARTO",
"T_PARTO",
"TIPO_EMB",
"CONS_EXP",
"IDPROFCER",
"N_HIJOSV",
"N_HIJOSM",
"EDAD_MADRE",
"CODPTORE",
"CODMUNRE",
"AREA_RES",
"SEG_SOCIAL",
"EST_CIVM",
"NIV_EDUM",
"ULTCURMAD",
"ASIS_MED",
"SEXO",
"T_GES",
"T_GES_AGRU_CIE",
"PESO_NAC"])

columnasFetal2017 = [col for col in fetal2017.columns if col in variablesABuscar]
columnasFetal2018 = [col for col in fetal2018.columns if col in variablesABuscar]
columnasFetal2021 = [col for col in fetal2021.columns if col in variablesABuscar]
columnasFetal2022 = [col for col in fetal2022.columns if col in variablesABuscar]

columnasFetal2017.sort()
columnasFetal2018.sort()
columnasFetal2021.sort()
columnasFetal2022.sort()

print(f"Las columnas de salud2017 son: {columnasFetal2017}")
print(f"Las columnas de salud2018 son: {columnasFetal2018}")
print(f"Las columnas de salud2021 son: {columnasFetal2021}")
print(f"Las columnas de salud2022 son: {columnasFetal2022}")

columnasEnComun = set(columnasFetal2017).intersection(set(columnasFetal2018)).intersection(set(columnasFetal2021)).intersection(set(columnasFetal2022))

print(f"Las columnas en común son: {columnasEnComun}")
print(f"Las columnas que no estan en todos los datasets son: {variablesABuscar - columnasEnComun}")
```

Al revisar los diccionarios de datos se puede apreciar que los atributos 'PMAN_MUER' y 'MAN_MUER' de los años 2017 y 2018, se unieron en el atributo P_PMAN_IRIS para los años 2021 y 2022. Asi que es necesario combinar estos dos atributos para los años 2017 y 2018

```python
columnasFetal2017.remove("MAN_MUER")
columnasFetal2018.remove("MAN_MUER")
columnasFetal2017.remove("PMAN_MUER")
columnasFetal2018.remove("PMAN_MUER")



columnasFetal2017.append("P_PMAN_IRIS")
columnasFetal2018.append("P_PMAN_IRIS")
columnasFetal2021.append("P_PMAN_IRIS")
columnasFetal2022.append("P_PMAN_IRIS")
```

```python
fetal2017[["PMAN_MUER","MAN_MUER"]]


import pandas as pd


def obtener_P_PMAN_IRIS(fila):
    if fila['PMAN_MUER'] == 1:
        return 0
    elif fila['MAN_MUER'] == 1:
        return 4
    elif fila['MAN_MUER'] == 2:
        return 1
    elif fila['MAN_MUER'] == 3:
        return 2
    elif fila['MAN_MUER'] == 4:
        return 6
    elif fila['MAN_MUER'] == 5:
        return 3
    return fila['PMAN_MUER']


fetal2017['PMAN_MUER'] = fetal2017.apply(obtener_P_PMAN_IRIS, axis=1)
fetal2018['PMAN_MUER'] = fetal2018.apply(obtener_P_PMAN_IRIS, axis=1)

fetal2017.rename(columns={"PMAN_MUER":"P_PMAN_IRIS"}, inplace=True)
fetal2018.rename(columns={"PMAN_MUER":"P_PMAN_IRIS"}, inplace=True)
```

Se puede ver que los atributos poseen valores validos salvo por dos de ellos : *ULTCURMAD* y *N_HIJOSM*, ya que ambos registran un valor de 0 cuando este no se encuentra dentro de sus posibles valores

Si bien los siguientes detalles no son problemas de validez, pueden aportar datos valiosos para futuros entregables:
- *MU_PARTO* nunca toma el valor de 3. 
- *N_HIJOSM* toma un valor maximo de 15, pero en los datos el maximo esta 14.
- *N_HIJOSV* dice que puede tomar un valor de 20 pero el maximo encontrado es 15.



A continuacion, se corrigen los errores presentados, para *ULTCURMAD*, ya que se cuenta con el valor "Sin información", se reemplazaran los 0's por 99's. Por el lado de *N_HIJOSM*, se reemplazaran los 0's usando la mediana



```python
fetal_df['ULTCURMAD'] = fetal_df['ULTCURMAD'].replace(0, 99)

mediana_hijosm = fetal_df['N_HIJOSM'].median()
fetal_df['N_HIJOSM'] = fetal_df['N_HIJOSM'].replace(0, mediana_hijosm)
```

Por ultimo, se verifica que *C_BAS1*, *COD_MUNIC*, *CODMUNRE* y tengan valores validos

```python
municipios_df = pd.read_csv("RegionesFormato.csv")
codigosMunicipio = set(municipios_df["CÓDIGO DANE DEL MUNICIPIO"].unique())


municipiosFetal = set(fetal_df["COD_MUNIC"].unique())
municipiosFallecido = set(fetal_df["CODMUNRE"].unique())


municipiosFaltantes = (municipiosFetal.union(municipiosFallecido)) - codigosMunicipio - set([np.nan])

print(len(municipiosFaltantes))
municipiosFaltantes
```

Con base en esto, podemos ver que todos los municipios son validos

```python
enfermedades_df = pd.read_excel("Cie.xls", skiprows=6)
codigosEnfermedades = set(enfermedades_df["COD_4"].unique())


enfermedadesFetal = set(fetal_df["C_BAS1"].unique())

enfermedadesFaltantes = enfermedadesFetal - codigosEnfermedades

print(len(enfermedadesFaltantes))
```

Existen varios problemas de consistencia al tratar a los valores nulos. A continuación se iran mencionando y corrigiendo
