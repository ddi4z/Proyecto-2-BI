## 1. Carga y Manipulación básica de los Datos

```python
# Importaciones de bibliotecas
import pandas as pd
```

```python
personas2017 = pd.read_csv("Personas2017.csv", delimiter=";")
personas2018 = pd.read_csv("Personas2018.csv", delimiter=";")
personas2021 = pd.read_csv("Personas2021.csv", delimiter=",")
personas2022 = pd.read_csv("Personas2022.csv", delimiter=",")

print(f"Las dimensiones de personas2017 son: {personas2017.shape}")
print(f"Las dimensiones de personas2018 son: {personas2018.shape}")
print(f"Las dimensiones de personas2021 son: {personas2021.shape}")
print(f"Las dimensiones de personas2022 son: {personas2022.shape}")
```

Las variables candidatas descritas en el informe son las siguientes:

- 1. Cabecera, 2. Resto (centros poblados y área rural dispersa) (Clase) 
- Cada una de las 24 a.M., otras cabeceras y resto (Dominio) 
- Dummy capital de departamento. 1: capital 0: no capital (Capital) 
- Estrato de energía para las 13 a.M., y sextil de icv para otras cabeceras y resto (Estrato1) 
- Sexo (P6020) 
- ¿Está afiliado, es cotizante o es beneficiario de alguna entidad de seguridad social en salud? (P6090) 
- ¿A cuál de los siguientes regímenes de seguridad social en salud está afiliado: (P6100) 
- ¿Cuál es el nivel educativo más alto alcanzado y el último año o grado aprobado en este nivel? (P6210) 
- Grado escolar aprobado (P6210s1) 
- Durante las últimas 4 semanas, ¿hizo diligencias para cambiar de trabajo? (P7150) 
- Población en edad de trabajar 1: sí 0: no (Pet) 
- Ocupado 1: sí (Oc) 


Se procederá a evidenciar si estos atributos se encuentran en los cuatro archivos de datos que se tienen disponibles.

```python
variablesABuscar = set(["Clase","Dominio","P6020", "P6090", "P6100", "P6210", "P6210s1", "P7150", "Pet", "Oc" ])

columnasPersonas2017 = [col.capitalize() for col in personas2017.columns if col.capitalize() in variablesABuscar]
columnasPersonas2018 = [col.capitalize() for col in personas2018.columns if col.capitalize() in variablesABuscar]
columnasPersonas2021 = [col.capitalize() for col in personas2021.columns if col.capitalize() in variablesABuscar]
columnasPersonas2022 = [col.capitalize() for col in personas2022.columns if col.capitalize() in variablesABuscar]

columnasPersonas2017.sort()
columnasPersonas2018.sort()
columnasPersonas2021.sort()
columnasPersonas2022.sort()

print(f"Las columnas de persona2017 son: {columnasPersonas2017}")
print(f"Las columnas de persona2018 son: {columnasPersonas2018}")
print(f"Las columnas de persona2021 son: {columnasPersonas2021}")
print(f"Las columnas de persona2022 son: {columnasPersonas2022}")

columnasEnComun = set(columnasPersonas2017).intersection(set(columnasPersonas2018)).intersection(set(columnasPersonas2021)).intersection(set(columnasPersonas2022))

print(f"Las columnas en común son: {columnasEnComun}")
print(f"Las columnas que no estan en todos los datasets son: {variablesABuscar - columnasEnComun}")
```

Con base en esto, se puede observar que no se encentran los siguientes atributos.


- Sexo (P6020) 
- ¿Cuál es el nivel educativo más alto alcanzado por y el último año o grado aprobado en este nivel? (P6210)
- Grado escolar aprobado (P6210s1) 

Fue posible encontrar el atributo sexo (p3271), el cual funge el mismo rol que P6020. Por lo tanto, se procederá a reemplazar P6020 por p3271 para el año 2022  
Fue posible encontrar el atributo ¿Cuál es el nivel educativo más alto alcanzado por y el último año o grado aprobado en este nivel? (p3042), el cual funge el mismo rol que P6210. Por lo tanto, se procederá a reemplazar P6210 por p3042 para el año 2022  
Fue posible encontrar el atributo Grado (p3042s1), el cual funge el mismo rol que P6210s1. Por lo tanto, se procederá a reemplazar P6210s1 por p3042s1 para el año 2022


```python
columnasPersonas2022.append("P3271")
columnasPersonas2022.append("P3042")
columnasPersonas2022.append("P3042s1")
```

Asimismo, para facilitiar su análisis, se procederá a cambiar sus nombres

```python
personas2017.columns = [col.capitalize() for col in personas2017.columns]
personas2018.columns = [col.capitalize() for col in personas2018.columns]
personas2021.columns = [col.capitalize() for col in personas2021.columns]
personas2022.columns = [col.capitalize() for col in personas2022.columns]

idANombre = {
    'P7150':'Durante las últimas 4 semanas, ¿hizo diligencias para cambiar de trabajo?',
    'P6090':'¿Está afiliado, es cotizante o es beneficiario de alguna entidad de seguridad social en salud?',
    'P6100':'¿A cuál de los siguientes regímenes de seguridad social en salud está afiliado?',
    'P6020': 'sexo',
    'P6210': '¿Cuál es el nivel educativo más alto alcanzado por y el último año o grado aprobado en este nivel?',
    'P6210s1': 'Grado',
    'P3271': 'sexo',
    'P3042': '¿Cuál es el nivel educativo más alto alcanzado por y el último año o grado aprobado en este nivel?',
    'P3042s1': 'Grado',
}
columnasPersonas2017 = [idANombre[col] if col in idANombre else col for col in columnasPersonas2017]
columnasPersonas2018 = [idANombre[col] if col in idANombre else col for col in columnasPersonas2018]
columnasPersonas2021 = [idANombre[col] if col in idANombre else col for col in columnasPersonas2021]
columnasPersonas2022 = [idANombre[col] if col in idANombre else col for col in columnasPersonas2022]



personas2017.rename(columns=idANombre, inplace=True)
personas2018.rename(columns=idANombre, inplace=True)
personas2021.rename(columns=idANombre, inplace=True)
personas2022.rename(columns=idANombre, inplace=True)
```

```python
for col in columnasPersonas2017:
    personas2017[col] = pd.to_numeric(personas2017[col], errors="coerce")

for col in columnasPersonas2018:
    personas2018[col] = pd.to_numeric(personas2018[col], errors="coerce")
```

Finalmente, se unen los 4 datasets en uno unico, para facilitar el analisis de los datos

```python
personas2017['año'] = 2017
personas2018['año'] = 2018
personas2021['año'] = 2021
personas2022['año'] = 2022

personas_df = pd.concat([personas2017[columnasPersonas2017 + ["año"]], personas2018[columnasPersonas2018+ ["año"]], personas2021[columnasPersonas2021+ ["año"]], personas2022[columnasPersonas2022+ ["año"]]], axis=0)
personas_df.reset_index(drop=True, inplace=True)
personas_df.sample(10)
```

## 2.1.1. Análisis de completitud
A continuación, se presentan los porcentajes de completitud e incompletitud para cada uno de los atributos.

```python
porcentajeIncompletitudAtributo =  (100 * personas_df.isnull().sum() / personas_df.shape[0]).sort_values(ascending=False)
pd.DataFrame({
    "Atributo": porcentajeIncompletitudAtributo.index,
    "Completitud (%)": 100 - porcentajeIncompletitudAtributo.values,
    "Incompletitud (%)": porcentajeIncompletitudAtributo.values
})
```

Si bien se ve una alta incompletitud en el atributo Durante las últimas 4 semanas, ¿hizo diligencias para cambiar de trabajo?  , se decidió no eliminarlo. Debido a que si bien es un porcentaje bajo, existe un gran numero de datos

## 2.1.2. Análisis de unicidad
A continuación, se buscará determinar cuántas filas duplicadas hay en el dataset

```python
registrosDuplicados = personas_df.duplicated()
totalDuplicados = registrosDuplicados.sum()
totalOpiniones = personas_df.shape[0]
porcentajeDuplicados = 100.0 * totalDuplicados / totalOpiniones
print(f"Se cuenta con {totalDuplicados} registros duplicados ({porcentajeDuplicados:.2f}%)")

```

Los datos presentan un alto porcentaje de duplicados. Sin embargo no se eliminaran debido a que tambien ofrecen información valiosa, ya que permitiran evidenciar patrones
